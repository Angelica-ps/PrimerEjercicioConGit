```python
import random

def simulador_vehiculo():
    
    try:
        distancia = int(input("Ingrese la distancia al obstaculo (en metros):"))
    except ValueError:
        print("¡Error! Ingresa un nuero valido.")
        return  
    velocidad = random.randint(10, 120)

    if distancia <= 20 and velocidad > 60:
        accion = "FRENADO DE EMERGENCIA"
    elif 21 <= distancia <= 50:
        accion = "DESACELERACION PREVENTIVA"
    else:
        accion = "MANTENER CURSO NORMAL"

    print("-----------------------------")
    print("REPORTE DEL VEHICULO AUTONOMO")
    print("-----------------------------")
    print(f"Distancia ingresada: {distancia} metros")
    print(f"Velocidad simulada: {velocidad} km/h")
    print(f"Accion del sistema: {accion}")
    print("----------------------------")


# Función menu
def menu_principal():
    print("--- SISTEMA DE CONTROL AUTOMOTRIZ ---")
    while True:
        print("1. Ejecutar prueba de frenado")
        print("2. Salir del programa")
        opcion = input("Seleccione una opcion:")

        if opcion == "1":
            simulador_vehiculo()
        elif opcion == "2":
            print("Saliendo del sistema...")
            break
        else:
            print("Opcion no valida. Intente nuevamente.")
            
if __name__=="__main__":
    menu_principal()
```
