```python
#Programa Evaluador de Creditos
#Demostrar la estructura de un arbol de decision anidado

#Score Crediticio
for score_cliente in range(400,700,50):
    deuda_mensual = 600
    ingreso_mensual = 2000
    #Proporcion de Ingresos comprometidos
    relacion_deuda= deuda_mensual /  ingreso_mensual
    print("Procesando cliente - Score", score_cliente,"Relacion deuda",relacion_deuda)
    #Nodo Raiz Clientes con Historial Excelente
    if score_cliente >= 700:
        #Rama A
        if relacion_deuda <= 0.35:
            print("Aprobado Automaticamente")
        else:
            print("Buen score sin embargo se encuentra muy endeudado")
    #Nodo Secundario Clientes con Historial Moderado
    elif score_cliente >=500:
        #Rama 8
        if relacion_deuda <= 0.20:
             print("Revision manual por parte del personal","\n","Score medio, bajo riesgo de deuda")
        else:
            print("Score medio y alta deuda")
    else:
        print("Rechazado Automaticamente, Riesgo financiero critico")
    
```

```python
#El programa va determinar si la contraseña es segura
#Reglas:Minimo 8 caracteres, al menos una mayuscula, al menos un numero
#Su contraseña es segura
#Su contraseña le falta seguridad - No cumple los 8 caracteres - No cumple almenos una mayuscula
# No cumple al menos un numero.

print("=== Verificador de seguridad de contraseña ===")

contrasena = input("Ingrese su contraseña: ")

# Verificar longitud mínima de 8 caracteres
longitud = len(contrasena) >= 8

# Si tiene al menos una mayúscula
tiene_mayuscula = False
for caracter in contrasena:
    if caracter.isupper():
        tiene_mayuscula = True
        break  # Ya encontramos una, no hace falta seguir revisando

# Si tiene al menos un numero
tiene_numero = False
for caracter in contrasena:
    if caracter.isdigit():
        tiene_numero = True
        break

# Evaluar todas las condiciones
if longitud and tiene_mayuscula and tiene_numero:
    print("Su contraseña es segura")
else:
    print("Le falta seguridad a su contraseña")
    if not longitud:
        print("No cumple con los 8 caracteres mínimos")
    if not tiene_mayuscula:
        print("No cumple con al menos una letra mayúscula")
    if not tiene_numero:
        print("No cumple con al menos un número")
```
