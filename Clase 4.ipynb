```python
opcion = 0

#Precios
precio_encebollado = 3.00
precio_empanada = 1.50
total_cuenta = 0.00
#Controla que el descuento se use solo 1 vez
descuento_aplicado = False 

while (opcion != 5):
    print("**** BIENVENIDO A LA CAJA REGISTRADORA ****")
    print("1. Agregar Encebollado ($3.00)")
    print("2. Agregar Empanada ($1.50)")
    print("3. Aplicar Descuento (10%)")
    print("4. Ver Cuenta Actual")
    print("5. Pagar y Salir")
    opcion = int(input("Elige una opción: "))

    if opcion == 1:
        print("----- Agregar Encebollado -----")
        unidades = int(input("¿Cuántas unidades?: "))
        total_cuenta = total_cuenta + (unidades * precio_encebollado)
        print("--->", unidades, "Encebollado(s) agregado(s).")

    elif opcion == 2:
        print("----- Agregar Empanada -----")
        unidades = int(input("¿Cuántas unidades?: "))
        total_cuenta = total_cuenta + (unidades * precio_empanada)
        print("--->", unidades, "Empanada(s) agregado(s).")

    elif opcion == 3:
        print("----- Aplicar Descuento -----")
        if descuento_aplicado == True:
            print("¡Descuento ya aplicado anteriormente!")
        elif total_cuenta <= 10.00:
            print(" No se puede aplicar: la cuenta debe ser mayor a $10.00")
        else:
            total_cuenta = total_cuenta * 0.90      # Aplica el 10% de descuento
            descuento_aplicado = True
            print(" ¡Descuento del 10% aplicado con éxito!")

    elif opcion == 4:
        print("----- Ver Cuenta Actual -----")
        print("Total a cobrar: $", total_cuenta)

    elif opcion == 5:
        print("----- Pagar y Salir -----")
        print("Total a cobrar: $", total_cuenta)
        efectivo = float(input("¿Efectivo recibido?: "))
        if efectivo < total_cuenta:
            print(" Dinero insuficiente, complete el pago.")
        else:
            vuelto = efectivo - total_cuenta
            print(" Pago exitoso. Su vuelto es: $", vuelto)
            print(" ¡Gracias por su compra, cerrando sistema!")

    else:
        print("Opcion Invalida, Selecciona el numero 1,2,3,4,5")

    
```
