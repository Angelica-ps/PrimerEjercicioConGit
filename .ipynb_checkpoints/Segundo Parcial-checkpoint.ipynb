```python
#Programa Evaluador de Creditos
#Demostrar la estructura de un arbol de decision anidado

#Score Crediticio
for score_cliente in range(400,700,50):
    deuda_mensual = 600
    ingreso_mensual = 2000
    #Proporcion de Ingresos comprometidos
    relacion_deuda= deuda_mensual /  ingreso_mensual
    print("Procesando cliente - Score", score_cliente,"Relacion deuda",relacion_deuda)
    #Nodo Raiz Clientes con Historial Excelente
    if score_cliente >= 700:
        #Rama A
        if relacion_deuda <= 0.35:
            print("Aprobado Automaticamente")
        else:
            print("Buen score sin embargo se encuentra muy endeudado")
    #Nodo Secundario Clientes con Historial Moderado
    elif score_cliente >=500:
        #Rama 8
        if relacion_deuda <= 0.20:
             print("Revision manual por parte del personal","\n","Score medio, bajo riesgo de deuda")
        else:
            print("Score medio y alta deuda")
    else:
        print("Rechazado Automaticamente, Riesgo financiero critico")
    
```

```python
#El programa va determinar si la contraseña es segura
#Reglas:Minimo 8 caracteres, al menos una mayuscula, al menos un numero
#Su contraseña es segura
#Su contraseña le falta seguridad - No cumple los 8 caracteres - No cumple almenos una mayuscula
# No cumple al menos un numero.

print("=== Verificador de seguridad de contraseña ===")

contrasena = input("Ingrese su contraseña: ")

# Verificar longitud mínima de 8 caracteres
longitud = len(contrasena) >= 8

# Si tiene al menos una mayúscula
tiene_mayuscula = False
for caracter in contrasena:
    if caracter.isupper():
        tiene_mayuscula = True
        break  # Ya encontramos una, no hace falta seguir revisando

# Si tiene al menos un numero
tiene_numero = False
for caracter in contrasena:
    if caracter.isdigit():
        tiene_numero = True
        break

# Evaluar todas las condiciones
if longitud and tiene_mayuscula and tiene_numero:
    print("Su contraseña es segura")
else:
    print("Le falta seguridad a su contraseña")
    if not longitud:
        print("No cumple con los 8 caracteres mínimos")
    if not tiene_mayuscula:
        print("No cumple con al menos una letra mayúscula")
    if not tiene_numero:
        print("No cumple con al menos un número")
```

```python
#El sistema de acceso requiere verificar que el usuario es humano resolviendo una suma matematica aleatoria.
import random

#Numeros aleatorios entre 1 y 20
numero1 = random.randint(1, 20)
numero2 = random.randint(1, 20)

respuesta_correcta = numero1 + numero2
intentos = 3

while intentos > 0:
    try:
        print(f"Resuelve la suma: {numero1} + {numero2} = ?")
        respuesta = int(input("Ingresa tu respuesta:"))

        if respuesta == respuesta_correcta:
            print("Acceso concedido.")
            break
        else:
            intentos = intentos- 1
            if intentos > 0:
                print(f"Respuesta incorrecta. Te quedan {intentos} intentos")
            else:
                print("SISTEMA BLOQUEADO.")
    except ValueError:
        print(f"Error: Ingresa solo números enteros.")
```

```python
import re
import time
print("Bienvenido al sistema de registro ITQ")
print("El correo debe ser minuscula y terminar en @itq.edu.ec")

#Definir el patron mediante REGEX
patron_correo=r"^[a-z.]+@itq\.edu\.ec$"
while True:
    correo_ingresado= input("Ingresa tu correo Institucional:")

    if re.match(patron_correo,correo_ingresado):
        print("Verificando dominio...")
        time.sleep(1)
        progreso=0
        
        while progreso<=100:
          print(f"\rRegistrando usuario {progreso}%",end="")
          time.sleep(0.2)
          progreso = progreso + 25

        print("Correo Aceptado, pasando a verificar edad....")
        break
    else:
        print(f"Error:Formato del correo invalido o contiene mayusculas")
        print(f"Intente de nuevo")
# Validación de edad
while True:
    try:
        edad = int(input("Ingresa tu edad:"))
        if edad >= 17:
            break
        else:
            print("Aviso:Debes ser mayor o igual a 17 años para registrarte.")
    except ValueError:
        print("Error:Ingresa solo números, no letras ni símbolos.")
 
print("Preparando registro...")
carga = 0
while carga <= 100:
    print(f"\rGenerando perfil... {carga}%", end="")
    time.sleep(0.3)
    carga += 20
print()
print(f"Bienvenido al sistema, usuario con correo: {correo_ingresado} ")
```

```python
#Boleteria de cine


def calcular_precio_cine(edad_cliente):
    precio_base = 7.50

    if edad_cliente < 12:
        return precio_base / 2
    elif edad_cliente >= 65:
        return precio_base -2.00
    else:
        return precio_base

edad = int(input("Ingrese la edad del cliente"))
total_pagar =calcular_precio_cine(edad)

print(f"El valor su entrada es ${total_pagar:.2f}")
```
